```python
import detectron2
import torch

print("Detectron2 OK")
print("CUDA disponível:", torch.cuda.is_available())
print("GPU:", torch.cuda.get_device_name(0))
```

```python
from detectron2.engine import DefaultPredictor
from detectron2.config import get_cfg
from detectron2 import model_zoo

cfg = get_cfg()
cfg.merge_from_file(
    model_zoo.get_config_file(
        "COCO-InstanceSegmentation/mask_rcnn_R_50_FPN_3x.yaml"
    )
)

cfg.MODEL.ROI_HEADS.SCORE_THRESH_TEST = 0.5
cfg.MODEL.WEIGHTS = model_zoo.get_checkpoint_url(
    "COCO-InstanceSegmentation/mask_rcnn_R_50_FPN_3x.yaml"
)

predictor = DefaultPredictor(cfg)

print("Modelo carregado com sucesso!")
```

```python
!wget https://upload.wikimedia.org/wikipedia/commons/6/60/Naxos_Taverna.jpg -O test.jpg
```

```python
import cv2
from detectron2.utils.visualizer import Visualizer
from detectron2.data import MetadataCatalog

img = cv2.imread("test.jpg")

outputs = predictor(img)

v = Visualizer(
    img[:, :, ::-1],
    MetadataCatalog.get(cfg.DATASETS.TRAIN[0]) if len(cfg.DATASETS.TRAIN) else None,
    scale=1.2
)

out = v.draw_instance_predictions(outputs["instances"].to("cpu"))
cv2.imwrite("resultado.jpg", out.get_image()[:, :, ::-1])

print("Inferência concluída")
```

```python
import json
import os

# caminhos
INPUT_JSON = "/content/carros_motos_pessoas.v1i.coco/train/_annotations.coco.json"
OUTPUT_JSON = "/content/carros_motos_pessoas.v1i.coco/train/annotations_clean.json"

# classes desejadas
CLASSES_OK = ["carro", "moto", "pessoa"]

with open(INPUT_JSON, "r") as f:
    coco = json.load(f)

# mapear categorias válidas
new_categories = []
old_id_to_new_id = {}
new_id = 0

for cat in coco["categories"]:
    name = cat["name"]
    if name == "motos":
        name = "moto"  # corrigir plural
    if name in CLASSES_OK:
        old_id_to_new_id[cat["id"]] = new_id
        new_categories.append({
            "id": new_id,
            "name": name
        })
        new_id += 1

# filtrar annotations
new_annotations = []
for ann in coco["annotations"]:
    if ann["category_id"] in old_id_to_new_id:
        ann["category_id"] = old_id_to_new_id[ann["category_id"]]
        new_annotations.append(ann)

# salvar novo coco
coco["categories"] = new_categories
coco["annotations"] = new_annotations

with open(OUTPUT_JSON, "w") as f:
    json.dump(coco, f)

print("JSON limpo criado com sucesso!")
```

```python
#CONFIGURAR MODELO
from detectron2.config import get_cfg
from detectron2 import model_zoo
import os

cfg = get_cfg()
cfg.merge_from_file(
    model_zoo.get_config_file(
        "COCO-Detection/faster_rcnn_R_50_FPN_3x.yaml"
    )
)

# Datasets
cfg.DATASETS.TRAIN = ("carros_motos_pessoas_train",)
cfg.DATASETS.TEST = ("carros_motos_pessoas_valid",)

cfg.DATALOADER.NUM_WORKERS = 2

# Pesos pré-treinados
cfg.MODEL.WEIGHTS = model_zoo.get_checkpoint_url(
    "COCO-Detection/faster_rcnn_R_50_FPN_3x.yaml"
)

# Treinamento
cfg.SOLVER.IMS_PER_BATCH = 2
cfg.SOLVER.BASE_LR = 0.00025
cfg.SOLVER.MAX_ITER = 1500   # bom para ~165 imagens
cfg.SOLVER.STEPS = []

# Classes
cfg.MODEL.ROI_HEADS.NUM_CLASSES = 3
cfg.MODEL.ROI_HEADS.SCORE_THRESH_TEST = 0.5

# Saída
cfg.OUTPUT_DIR = "/content/output"
os.makedirs(cfg.OUTPUT_DIR, exist_ok=True)
```

```python
#REINICIAR TREINAMENTO
from detectron2.engine import DefaultTrainer

trainer = DefaultTrainer(cfg)
trainer.resume_or_load(resume=False)
trainer.train()
```
